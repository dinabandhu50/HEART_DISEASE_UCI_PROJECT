### Separating categorical from numerical variables

```python
cat_var =[]
num_var =[]
for item in df.columns:
    ln = len(df[item].unique())
    if ln<7:
        cat_var.append(item)
    else:
        num_var.append(item)
print(cat_var)
print(num_var+['target'])
```

```python
g1 = sns.pairplot(df,hue='target',vars=num_var)
g1.fig.set_size_inches(15,15)
```

```python
g2 =sns.pairplot(df,hue='target',vars=cat_var)
g2.fig.set_size_inches(15,15)
```

# Model training

### train test split

```python
X = df.drop(['target'],axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

### Model building

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='lbfgs',max_iter=5000)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

# Model evaluation

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

# Model saving

```python
import pickle
```

```python
PATH_MODEL = './models/model_saved.pkl'
with open(PATH_MODEL,'wb') as fid:
    pickle.dump(model,fid)
```
